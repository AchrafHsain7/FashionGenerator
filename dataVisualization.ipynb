```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Load the Data

```python
labelsTitles = ["T-shirt","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]
```

```python
data = pd.read_csv("fashionMNIST/fashion-mnist_train.csv")
train_labels = data.pop("label").to_numpy()
train_np = data.to_numpy(dtype=np.float64)
train_np = train_np.reshape((-1, 28, 28))
train_np /= 255.0
print(train_np.shape)
del data
```

```python
data = pd.read_csv("fashionMNIST/fashion-mnist_test.csv")
test_labels = data.pop("label").to_numpy()
test_np = data.to_numpy(dtype=np.float64)
test_np = test_np.reshape((-1, 28, 28))
test_np /= 255.0
print(test_np.shape)
del data
```

# Visualization

```python
start = 200
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(train_np[i+start])
    plt.title(labelsTitles[train_labels[i+start]])

plt.tight_layout()
plt.show()
```

# Savin Data

```python
np.savez_compressed("data/train.npz", x=train_np, y=train_labels)
np.savez_compressed("data/test.npz", x=test_np, y=test_labels)
```
